```python
def thong_ke_ky_tu(line):
    line = line.replace('\n', '')
    demkt = dict({})
    for c in line:
        if demkt.get(c) is None:
            demkt[c] = 1
        else:
            demkt[c] = demkt[c] + 1
        pass

    # print(demkt)
    return demkt
    pass

# thong_ke_ky_tu(lines[0])
```

```python
def thong_ke_tu(line):
    line = line.replace('\n', '') # stopwords
    line = line.replace('.', '')
    # print(line)
    ds_tu = line.split(' ')
    # print(ds_tu)
    ds_tu = [w.lower() for w in ds_tu]
    # print(ds_tu)

    demtu_dict = dict({})
    for w in ds_tu:
        if demtu_dict.get(w) is None:
            demtu_dict[w] = 1
        else:
            demtu_dict[w] = demtu_dict[w] + 1
    # print(demtu_dict)
    return demtu_dict
    pass

# thong_ke_tu(lines[1])
```
